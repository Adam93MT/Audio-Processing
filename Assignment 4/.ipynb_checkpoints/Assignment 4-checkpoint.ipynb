```python
%pylab inline
import numpy as np
from scipy import io
from scipy import signal
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-notebook')
```

```python
def normalize(x):
    return x/float(np.max(np.abs(x)))
def dB(x):
    return 20 * np.log10(x)
```

# Assignment \#4
 
## Part1:
 
For each of the two files, Oboe.wav  and Trumpet.wav
Assume the fundamental frequency is f0=233.8Hz (Bb3).
 
Construct and apply the following filters, all Fourth Order butterworth:

- LPF (low pass filter):  Cuttoff frequency 4*f0

- BPF (band pass filter): 2/3 f0 to 4/3*f0

- HPF (high pass filter):  2*f0
 
You should submit 6 (six) wave files called:
OboeLPF.wav
OboeBPF.wav
OboeHPF.wav
and the same for Trumpet.

```python
f0 = 233.8 #Hz
w0 = f0*2*np.pi

[fs, oboe] = io.wavfile.read('Oboe.wav')
[fs, trumpet] = io.wavfile.read('Trumpet.wav')

oboe = oboe/float(np.max(np.abs(oboe)))
trumpet = trumpet/float(np.max(np.abs(trumpet)))

Ny = fs/float(2)

t_ob = np.linspace(0, len(oboe)/fs, len(oboe))
t_tr = np.linspace(0, len(trumpet)/fs, len(trumpet))
# print t_tr.shape
# print trumpet.shape

OBOE = np.abs(np.fft.fftshift(np.fft.fft(oboe)))
TRUMPET = np.abs(np.fft.fftshift(np.fft.fft(trumpet)))

# np.fft.fftshift
f_ob = np.linspace(-Ny, Ny, len(oboe))
f_tr = np.linspace(-Ny, Ny, len(trumpet))

w_ob = np.linspace(-Ny, Ny, len(oboe)) * 2*np.pi
w_tr = np.linspace(-Ny, Ny, len(trumpet)) * 2*np.pi

figure()
figure(figsize=(14,3))
subplot(1,2,1)
plot(t_ob, oboe)
title('Oboe')
xlabel('$t$')
subplot(1,2,2)
xlabel('Hz')
semilogx(f_ob, OBOE);

figure()
figure(figsize=(14,3))
subplot(1,2,1)
title('Trumpet')
xlabel('$t$')
plot(t_tr, trumpet)
subplot(1,2,2)
xlabel('Hz')
semilogx(f_tr, TRUMPET);
```

```python
[LPb, LPa] = signal.butter(4, 4*f0/Ny, btype='low', analog=False)
[BPb, BPa] = signal.butter(4, [2/float(3)*f0/Ny, 4/float(3)*f0/Ny], btype='band', analog=False)
[HPb, HPa] = signal.butter(4, 2*f0/Ny, btype='high', analog=False)

print "LPF: " + str(4*f0) + "Hz"
print "BPF: " + str(2/float(3)*f0) + "Hz, " + str(4/float(3)*f0) + "Hz"
print "HPF: " + str(2*f0) + "Hz"

w_lp, h_lp = signal.freqz(LPb, LPa)
w_bp, h_bp = signal.freqz(BPb, BPa)
w_hp, h_hp = signal.freqz(HPb, HPa)

figure()
semilogx(w_lp, dB(h_lp))
semilogx(w_bp, dB(h_bp))
semilogx(w_hp, dB(h_hp))
xlabel("Normalized Frequency")
ylabel("dB")
grid(True);
```

```python
# Apply the filters to the TIME domain signal
oboe_lpf = normalize(signal.lfilter(LPb, LPa, oboe))
oboe_bpf = normalize(signal.lfilter(BPb, BPa, oboe))
oboe_hpf = normalize(signal.lfilter(HPb, HPa, oboe))

trumpet_lpf = normalize(signal.lfilter(LPb, LPa, trumpet))
trumpet_bpf = normalize(signal.lfilter(BPb, BPa, trumpet))
trumpet_hpf = normalize(signal.lfilter(HPb, HPa, trumpet))

# Calculate the FFT to verify
OBOE_LPF = np.abs(np.fft.fftshift(np.fft.fft(oboe_lpf)))
OBOE_BPF = np.abs(np.fft.fftshift(np.fft.fft(oboe_bpf)))
OBOE_HPF = np.abs(np.fft.fftshift(np.fft.fft(oboe_hpf)))

TRUMPET_LPF = np.abs(np.fft.fftshift(np.fft.fft(trumpet_lpf)))
TRUMPET_BPF = np.abs(np.fft.fftshift(np.fft.fft(trumpet_bpf)))
TRUMPET_HPF = np.abs(np.fft.fftshift(np.fft.fft(trumpet_hpf)))
```

```python
figure()
figure(figsize=(14,3))
subplot(1,2,1)
plot(t_ob, oboe_lpf)
title('Oboe LPF')
xlabel('$t$')
# xlim(0,0.5)
subplot(1,2,2)
xlabel('Hz')
semilogx(f_ob, OBOE);
semilogx(f_ob, OBOE_LPF)
plt.axvline(4*f0, color='green') # cutoff frequency
xlim(10,Ny);

figure()
figure(figsize=(14,3))
subplot(1,2,1)
plot(t_ob, oboe_bpf)
title('Oboe BPF')
xlabel('$t$')
# xlim(0,0.5)
subplot(1,2,2)
xlabel('Hz')
semilogx(f_ob, OBOE)
semilogx(f_ob, OBOE_BPF)
plt.axvline(2/float(3)*f0, color='green') # cutoff frequency;
plt.axvline(4/float(3)*f0, color='green') # cutoff frequency
xlim(10,Ny);

figure()
figure(figsize=(14,3))
subplot(1,2,1)
plot(t_ob, oboe_hpf)
title('Oboe HPF')
xlabel('$t$')
# xlim(0,0.5)
subplot(1,2,2)
xlabel('Hz')
semilogx(f_ob, OBOE)
semilogx(f_ob, OBOE_HPF)
plt.axvline(2*f0, color='green') # cutoff frequency
xlim(10,Ny);

t_tr = np.linspace(0, len(TRUMPET_LPF)/fs, len(TRUMPET_LPF))

figure()
figure(figsize=(14,3))
subplot(1,2,1)
plot(t_tr, trumpet_lpf)
title('Trumpet LPF')
xlabel('$t$')
# xlim(0,0.5)
subplot(1,2,2)
xlabel('Hz')
semilogx(f_tr, TRUMPET)
semilogx(f_tr, TRUMPET_LPF)
plt.axvline(4*f0, color='green') # cutoff frequency
xlim(10,Ny);

figure()
figure(figsize=(14,3))
subplot(1,2,1)
plot(t_tr, trumpet_bpf)
title('Trumpet BPF')
xlabel('$t$')
# xlim(0,0.5)
subplot(1,2,2)
xlabel('Hz')
semilogx(f_tr, TRUMPET)
semilogx(f_tr, TRUMPET_BPF)
plt.axvline(2/float(3)*f0, color='green') # cutoff frequency;
plt.axvline(4/float(3)*f0, color='green') # cutoff frequency;;
xlim(10,Ny);

figure()
figure(figsize=(14,3))
subplot(1,2,1)
plot(t_tr, trumpet_hpf)
title('Trumpet HPF')
xlabel('$t$')
# xlim(0,0.5)
subplot(1,2,2)
xlabel('Hz')
semilogx(f_tr, TRUMPET)
semilogx(f_tr, TRUMPET_HPF)
plt.axvline(2*f0, color='green') # cutoff frequency
xlim(10,Ny);
```

```python
io.wavfile.write('OboeLPF.wav', fs, oboe_lpf)
io.wavfile.write('OboeBPF.wav', fs, oboe_bpf)
io.wavfile.write('OboeHPF.wav', fs, oboe_hpf)

io.wavfile.write('TrumpetLPF.wav', fs, trumpet_lpf)
io.wavfile.write('TrumpetBPF.wav', fs, trumpet_bpf)
io.wavfile.write('TrumpetHPF.wav', fs, trumpet_hpf)
```

## Part 2
Load the following Stereo WAV file: Asst4Q2.wav

The file contains 16383 points.

The left channel is the input (an MLS sequence).

The right channel is the response of an equalizer.
 
- Compute the DFT of the right channel.

- Denote the right channel y(n), n=1..16383.

- Compute the DFT, Y=fft(y).  By inspection you should be able to read off the Peaks and Troughs in the equalizer response.

- Fill in the values in the following table.

- Note: these are standard 1/3 Octave frequencies.  All slots are either zero, +12dB or -12dB.

- Fill in the table below with either 0 (no change) + (boosted 12dB) or - (reduced 12dB).
 
Note: most are unchanged!
 
25	31.5	40	50	63	80	100	125	160	200
 
250	315	400	500	630	800	1000	1250	1600	2000
 
2500	3150	4000	5000	6300	8000	10000	12500	16000	20000


```python
[fs, a4q2] = io.wavfile.read('Asst4Q2.wav')
t = np.linspace(0, len(a4q2)/float(fs), len(a4q2))

left = a4q2[:,0]
right = a4q2[:,1]

figure()
plot(t, right)
plot (t, left);
```

```python
LEFT = np.abs(np.fft.fftshift(np.fft.fft(left)))
RIGHT = np.abs(np.fft.fftshift(np.fft.fft(right)))
Ny = fs/float(2)
N = len(left)
f = np.linspace(-Ny, Ny, N)

harmonix = [25,31.5,40,50,63,80,100,125,160,200,250,315,400,500,630,800,1000,1250,1600,2000, 2500,3150,4000,5000,6300,8000,10000,12500,16000,20000]
idx0 = int(len(left)/float(2))
harm_idx = idx0 + np.dot(harmonix,N/(2*Ny))

R = np.zeros(len(harm_idx))
L = np.zeros(len(harm_idx))
DIFF = np.zeros(len(harm_idx))
for i,idx in enumerate(harm_idx):
    idx = int(idx)
    R[i] = dB(RIGHT[idx])
    L[i] = dB(LEFT[idx])
    DIFF[i] = dB(RIGHT[idx]) - dB(LEFT[idx])

figure()
semilogx(f, dB(RIGHT))
semilogx(f, dB(LEFT))
scatter((harmonix), R)
scatter((harmonix), L)
# scatter((harmonix), DIFF)
xlabel('Freq (Hz)')
ylabel('dB');
```

```python
figure()
semilogx(f,dB(RIGHT)-dB(LEFT), color="green")
scatter((harmonix), DIFF, color="green")
xlabel('Freq (Hz)')
ylabel('dB');

for i,d in enumerate(DIFF):
    if abs(d) > 10:
        DIFF[i] = np.sign(d)*12
    else:
        DIFF[i] = 0

df = pd.DataFrame({"Octave Band":harmonix, "Difference":DIFF})
print df;

with open('results.txt','wb') as outfile:
    ostring = str(df)
    outfile.write(ostring.encode('utf-8'))
```
